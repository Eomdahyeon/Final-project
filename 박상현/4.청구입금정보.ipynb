```python
data_type = "train"
# month = "07"
# category = "청구입금정보"

# local
root_path = '../data/open'

# colab
# root_path = '/content/drive/MyDrive/12조 파이널프로젝트/data'

drive_folder = f'{root_path}/{data_type}/4.청구입금정보/'
```

```python
# parquet 파일 데이터를 읽어온다.
file_name = pd.read_parquet(f'{drive_folder}201807_train_청구정보.parquet')
file_name
```

```python
# 1. 파일 경로 패턴 설정
file_list = glob.glob(f'{drive_folder}/2018*_train_청구정보.parquet')

# 2. 각 파일 읽어서 리스트에 담기
df_list = [pd.read_parquet(file) for file in file_list]

# 3. 데이터프레임 하나로 합치기
df_merged = pd.concat(df_list, ignore_index=True)

# 4. csv으로 저장 -> 필요시 parquet으로 가능
df_merged.to_csv(f'{drive_folder}/train_청구정보_통합.csv', index=False)

print("✅ 총 파일 수:", len(file_list))
print("✅ 통합 완료! 저장 파일명: train_청구정보_통합.csv")
print("✅ 총 행 수:", len(df_merged))
```
