```python
# 메모리 사용량이 너무 많아서, float64를 float32로 바꾼다.
for col in performance.select_dtypes(include='float64').columns:
    performance[col] = performance[col].astype('float32')
```

```python
# 유일값 개수가 1개 이하인 컬럼 찾기 (전부 0이거나 전부 NaN)
constant_cols = [col for col in performance.columns if performance[col].nunique(dropna=False) <= 1]

print("🧊 모든 값이 동일한 컬럼 (전부 0이거나 한 값만 있음):")
print(constant_cols)
```

```python
# 혜택수혜율_B0M, 혜택수혜율_R3M 둘다 0이 압도적으로 많음.
# 0과 99.999802로 양극단에 몰려있음.
# 결측치를 중앙값으로 채우는게 합리적으로 보임.

performance['혜택수혜율_B0M'] = performance['혜택수혜율_B0M'].fillna(performance['혜택수혜율_B0M'].median())
performance['혜택수혜율_R3M'] = performance['혜택수혜율_R3M'].fillna(performance['혜택수혜율_R3M'].median())
```
