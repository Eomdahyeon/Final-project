```python
data_type1 = "train"
data_type2 = "test"
# month = "07"
# category = "성과정보"

# local
root_path = '../data/open'

# colab
# root_path = '/content/drive/MyDrive/12조 파이널프로젝트/data'

drive_folder1 = f'{root_path}/{data_type1}/8.성과정보/'
drive_folder2 = f'{root_path}/{data_type2}/8.성과정보/'
```

```python
# parquet 파일 데이터를 읽어온다.
df1 = pd.read_parquet(f'{drive_folder1}201807_train_성과정보.parquet')
df1
```

```python
# 1. 파일 경로 패턴 설정
file_list = glob.glob(f'{drive_folder1}/2018*_train_성과정보.parquet')

# 2. 각 파일 읽어서 리스트에 담기
df_list = [pd.read_parquet(file) for file in file_list]

# 3. 데이터프레임 하나로 합치기 // ignore_index=True 를 이용해 인덱스 겹치지 않게
df_merged = pd.concat(df_list, ignore_index=True)

# 4. parquet로 저장
df_merged.to_parquet(f'{drive_folder1}train_성과정보_통합.parquet', index=False)

print("✅ 총 파일 수:", len(file_list))
print("✅ 통합 완료! 저장 파일명: train_성과정보_통합.parquet")
print("✅ 총 행 수:", len(df_merged))
```
