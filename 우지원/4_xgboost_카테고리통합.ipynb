```python
data_splits = ["train","test"]
months = ["07","08","09","10","11","12"]
# ê° ë°ì´í„° ìœ í˜•ë³„ í´ë”ëª…, íŒŒì¼ ì ‘ë¯¸ì‚¬, ë³€ìˆ˜ ì ‘ë‘ì–´ ì„¤ì •
data_categories = {
    "íšŒì›ì •ë³´": {"folder": "1.íšŒì›ì •ë³´", "suffix": "íšŒì›ì •ë³´", "var_prefix": "customer"},
    "ì‹ ìš©ì •ë³´": {"folder": "2.ì‹ ìš©ì •ë³´", "suffix": "ì‹ ìš©ì •ë³´", "var_prefix": "credit"},
    "ìŠ¹ì¸ë§¤ì¶œì •ë³´": {"folder": "3.ìŠ¹ì¸ë§¤ì¶œì •ë³´", "suffix": "ìŠ¹ì¸ë§¤ì¶œì •ë³´", "var_prefix": "sales"},
    "ì²­êµ¬ì •ë³´": {"folder": "4.ì²­êµ¬ì…ê¸ˆì •ë³´", "suffix": "ì²­êµ¬ì •ë³´", "var_prefix": "billing"},
    "ì”ì•¡ì •ë³´": {"folder": "5.ì”ì•¡ì •ë³´", "suffix": "ì”ì•¡ì •ë³´", "var_prefix": "balance"},
    "ì±„ë„ì •ë³´": {"folder": "6.ì±„ë„ì •ë³´", "suffix": "ì±„ë„ì •ë³´", "var_prefix": "channel"},
    "ë§ˆì¼€íŒ…ì •ë³´": {"folder": "7.ë§ˆì¼€íŒ…ì •ë³´", "suffix": "ë§ˆì¼€íŒ…ì •ë³´", "var_prefix": "marketing"},
    "ì„±ê³¼ì •ë³´": {"folder": "8.ì„±ê³¼ì •ë³´", "suffix": "ì„±ê³¼ì •ë³´", "var_prefix": "performance"}
}

### âš ï¸ ì´ ë¶€ë¶„ ê¼­ í™•ì¸í•´ì„œ ì„¤ì •í•œ ë’¤ì— ì‹¤í–‰!!
# local
root_path = '../data/open'

# colab
# root_path = '/content/drive/MyDrive/12á„Œá…© á„‘á…¡á„‹á…µá„‚á…¥á†¯á„‘á…³á„…á…©á„Œá…¦á†¨á„á…³/data'
```

```python
### âš ï¸ ì´ ë¶€ë¶„ ê¼­ í™•ì¸í•´ì„œ ì„¤ì •í•œ ë’¤ì— ì‹¤í–‰!!
# ë¨¸ì‹ ëŸ¬ë‹ ëŒë¦´ ì›” ì„ íƒ 
select_month_num = 5
```

### XGBoost í™œìš©í•´ì„œ ëª¨ë¸ í•™ìŠµ ë° ì˜ˆì¸¡

```python
### 1. íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
# í›ˆë ¨ ë°ì´í„°
file_name=f"{root_path}/{data_splits[0]}/2018{months[select_month_num]}_processed_All.parquet"
train_df=pd.read_parquet(file_name)
# í…ŒìŠ¤íŠ¸ ë°ì´í„°
test_file =f"{root_path}/{data_splits[1]}/2018{months[select_month_num]}_processed_All.parquet"
test_df = pd.read_parquet(test_file)

### 2. ê³µí†µ í”¼ì²˜ ì¶”ì¶œìš© ì„¤ì •
target_cols = ['Segment_A', 'Segment_B', 'Segment_C', 'Segment_D', 'Segment_E']
exclude_cols = ['ID', 'ê¸°ì¤€ë…„ì›”'] + target_cols
# object íƒ€ì… ì»¬ëŸ¼ë„ ì œì™¸í•˜ë©´ì„œ feature ì»¬ëŸ¼ êµ¬ì„±
feature_cols = [
    col for col in train_df.columns
    if col not in exclude_cols and train_df[col].dtype != 'object'
]

### 3. ê²°ê³¼ ì €ì¥ìš©
all_results = {}

for target in target_cols:
    print(f"\nğŸ”· íƒ€ê²Ÿ: {target}")
    
    # í•™ìŠµ ë°ì´í„°
    X_train = train_df[feature_cols]
    y_train = train_df[target]
    
    # í…ŒìŠ¤íŠ¸ ë°ì´í„° (ë™ì¼í•œ feature ì»¬ëŸ¼ë§Œ ì‚¬ìš©)
    # ê·¸ ì¤‘ ì‹¤ì œ test_dfì— ì¡´ì¬í•˜ëŠ” ì»¬ëŸ¼ë§Œ ì¶”ì¶œ
    valid_feature_cols = [col for col in feature_cols if col in test_df.columns]

    # ì ìš©
    X_test = test_df[valid_feature_cols]
    
    # 1ì°¨ ëª¨ë¸ë¡œ ì¤‘ìš”ë„ ì¶”ì¶œ
    model_full = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
    model_full.fit(X_train, y_train)
    
    importance = model_full.feature_importances_
    importance_df = pd.DataFrame({
        'feature': X_train.columns,
        'importance': importance
    }).sort_values(by='importance', ascending=False)
    
    # ìƒìœ„ 30ê°œ í”¼ì²˜ ì„ íƒ
    top_features = importance_df.head(30)['feature'].tolist()
    
    # ìƒìœ„ í”¼ì²˜ë§Œ ì¶”ë ¤ì„œ ë‹¤ì‹œ í•™ìŠµ
    model_top = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
    model_top.fit(X_train[top_features], y_train)

    # ì˜ˆì¸¡ í™•ë¥  ì €ì¥ (soft voting)
    y_proba = model_top.predict_proba(X_test[top_features])[:, 1]  # class 1 í™•ë¥ 
    test_df[f'proba_{target}'] = y_proba

    print(f"âœ… ì˜ˆì¸¡ ì™„ë£Œ â†’ test_df[f'proba_{target}'] ìƒì„±ë¨")

    # ì‹œê°í™”
    plt.figure(figsize=(10, 6))
    importance_df.head(30).plot.barh(x='feature', y='importance', legend=False)
    plt.title(f"Top 30 Feature Importance for {target}")
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
    
    # ê²°ê³¼ ì €ì¥
    all_results[target] = {
        'top_features': top_features,
        'importance_df': importance_df.head(30)
    }
```

```python
### 1. ê° íƒ€ê²Ÿë³„ ì˜ˆì¸¡ê°’ì€ test_df['pred_Segment_X'] ì— ì €ì¥ë˜ì–´ ìˆë‹¤ê³  ê°€ì •
segment_labels = ['A', 'B', 'C', 'D', 'E']
proba_cols = [f'proba_Segment_{label}' for label in segment_labels]

### 2. ê° rowì—ì„œ í™•ë¥ ì´ ê°€ì¥ ë†’ì€ ì»¬ëŸ¼ì˜ label ì„ íƒ
def get_top_segment(row):
    max_idx = row[proba_cols].values.argmax()
    return segment_labels[max_idx]

### 3. Segment ì´ë¦„ ë¶™ì´ê¸°
test_df['Segment'] = test_df.apply(get_top_segment, axis=1)

### 4. ID, Segmentë§Œ ì¶”ì¶œ
submission = test_df[['ID', 'Segment']]

### 5. ì €ì¥
output_file = f"../results/2018{months[select_month_num]}_All_xgboost_submission.csv"
submission.to_csv(output_file, index=False)

print("âœ… ì˜ˆì¸¡ ê²°ê³¼ê°€ csv íŒŒì¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```
