```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import f1_score
import xgboost as xgb

# ğŸ”¹ 1. ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
file_path = "/content/drive/MyDrive/12ì¡° íŒŒì´ë„í”„ë¡œì íŠ¸/data/train/201812_add_segment1.parquet"
df = pd.read_parquet(file_path)

# ğŸ”¹ 2. Feature / Target ì„¤ì •
X = df.drop(columns=['ID', 'ê¸°ì¤€ë…„ì›”','Segment1','Segment'], errors='ignore')
y = df['Segment1']  # 'E' ë˜ëŠ” 'other'

# ğŸ”¹ 3. ë¼ë²¨ ì¸ì½”ë”© (ë¬¸ì -> ìˆ«ì)
le = LabelEncoder()
y_encoded = le.fit_transform(y)  # 'E':0, 'other':1 (ë˜ëŠ” ë°˜ëŒ€)

# ğŸ”¹ 4. í•™ìŠµ/ê²€ì¦ ë°ì´í„° ë¶„ë¦¬
X_train, X_val, y_train, y_val = train_test_split(
    X, y_encoded,
    test_size=0.2,
    stratify=y_encoded,
    random_state=42
)

# ğŸ”¹ 5. XGBoost ëª¨ë¸ ì •ì˜ ë° í•™ìŠµ
model = xgb.XGBClassifier(
    objective='binary:logistic',
    use_label_encoder=False,
    random_state=42
    #tree_method='hist'  # Colabì—ì„œ ì†ë„ ë¹ ë¥´ê²Œ
)

model.fit(X_train, y_train)

# ğŸ”¹ 6. ì˜ˆì¸¡ ë° í‰ê°€
y_pred = model.predict(X_val)
f1 = f1_score(y_val, y_pred, average='micro')
print(f"\nâœ… [Validation F1 Score (micro)]: {f1:.4f}")
print("ğŸ“˜ í´ë˜ìŠ¤ ì¸ì½”ë”© ìˆœì„œ:", list(le.classes_))
```
