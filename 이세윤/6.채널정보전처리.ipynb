### 이상치 확인

```python
num_cols = df_채널정보.select_dtypes(include='int64').columns
```

```python
outlier_summary = []

for col in num_cols:
    Q1 = df_채널정보[col].quantile(0.25)
    Q3 = df_채널정보[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df_채널정보[(df_채널정보[col] < lower) | (df_채널정보[col] > upper)][col]
    outlier_count = outliers.count()
    outlier_ratio = outlier_count / len(df_채널정보)

    outlier_summary.append({
        '변수명': col,
        '이상치 수': outlier_count,
        '이상치 비율': round(outlier_ratio, 4),
        '최소값': df_채널정보[col].min(),
        '최대값': df_채널정보[col].max(),
        'Q1': Q1,
        'Q3': Q3
    })

outlier_df = pd.DataFrame(outlier_summary).sort_values(by='이상치 비율', ascending=False)
```
